### 1.1  Let's write a function which compute the t-th Fibonacci for any number:

#### Let's call this function Fibonacci(n)

```python
def Fibonacci(n):
    if n<=1:
        return n
    return Fibonacci(n-1) + Fibonacci(n-2)
```

### * A function that returns one realisation of the following random device:

Let's create the function "draw" 
Such that the payoff is 1 if there is "k consecutive successes" in a sequence

```python
from numpy.random import uniform

def draw(k):  

    payoff = 0
    count = 0

    for i in range(10):
        U = uniform()
        count = count + 1 if U < 0.5 else 0
        print(count)  
        if count >= k:
            payoff = 1

    return payoff

draw(3)
```

### * A function that does the same task except that the second rule of the above random device becomes:

Let's create the function "draw_new"  
Such that the payoff is 1 if there is "k successes" in a sequence

```python
def draw_new(k):  
    payoff = 0
    count = 0

    for i in range(10):
        U = uniform()
        count = count + ( 1 if U < 0.5 else 0 )
        print(count)
        if count >= k:
            payoff = 1

    return payoff
```

### 2. Computing the value of pi

```python
n=100000

count=0
for i in range (n):
    u, v= np.random.uniform() , np.random.uniform()
    d=np.sqrt((u-0.5)**2 + (v-0.5)**2)
    if d < 0.5:
        count += 1
        
area_estimate = count / n       

print("Then π is equal to", area_estimate * 4)
```
